```python
from sqlalchemy import create_engine
import psycopg2

# Load Excel file
df = pd.read_excel("movie_dataset.xlsx")
print(df.head())

# PostgreSQL credentials
host = "localhost"
user = "postgres"
password = "129141"
port = "5432"  # default PostgreSQL port
db_name = "movie_rating_system"

# Connect to default database to create new DB if not exists
conn = psycopg2.connect(host=host, user=user, password=password, dbname="postgres")
conn.autocommit = True  # Needed to CREATE DATABASE
cur = conn.cursor()

# Create database if it doesn't exist
cur.execute(f"SELECT 1 FROM pg_database WHERE datname='{db_name}';")
exists = cur.fetchone()
if not exists:
    cur.execute(f"CREATE DATABASE {db_name};")
    print(f"Database '{db_name}' created successfully.")
else:
    print(f"Database '{db_name}' already exists.")

cur.close()
conn.close()

# Connect to the new database using SQLAlchemy
engine = create_engine(f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{db_name}")

# Store DataFrame into a table named 'movies' (replace if exists)
df.to_sql('movies', engine, if_exists='replace', index=False)
print("DataFrame stored successfully in PostgreSQL table 'movies'.")
```

```python

# Read the table
df_movies = pd.read_sql("SELECT * FROM movies;", engine)
print(df_movies.head())
```
